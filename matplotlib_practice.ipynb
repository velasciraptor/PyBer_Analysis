```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
## LINE GRAPHS ##

# MATLAB METHOD

# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# OBJECT-ORIENTED METHOD
# Create the plot with ax.plt()
# fig, ax = plt.subplots()
# ax.plot(x_axis, y_axis)

# The plt.subplots() function returns a tuple that contains a figure and axes object(s).
# fig, ax = plt.subplots() unpacks the tuple into the variables fig and ax.
# The first variable, fig, references the figure.
# The second variable, ax, references the axes of the figure and allows us to annotate the figure.
# The ax.plot(x_axis, y_axis) plots the data from the x and y axes.
# The fig, ax = plt.subplots() line of code can also be written like this.

fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)

# or 
# Create the plot with ax.plt()
# ax = plt.axes()
# ax.plot(x_axis, y_axis)
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="D", color="green", linewidth=1, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
## BAR CHARTS ## MATLAB METHOD

# Create the plot.
plt.bar(x_axis, y_axis, color="green", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# HORIZONTAL BAR CHART

# Create a horizontal bar chart
plt.barh(x_axis, y_axis, color="magenta", label="Boston")
plt.title("PyBer Fare by Month")
plt.xlabel("Fare(s)")
plt.ylabel("Date")
plt.legend()
plt.gca().invert_yaxis()

# If you place something on the wrong axis, you can inveret using code below.
# gca = get current axes
# plt.barh(x_axis, y_axis)
# plt.gca().invert_yaxis()
```

```python
## BAR CHARTS ## OBJECT-ORIENTED METHOD

# BAR CHART
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, color="black", label="New York")
ax.legend()
```

```python
# HORIZONTAL BAR CHART

# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color="cyan", label="Chicago")
ax.legend()
ax.set_title("PyBer Fare by month")
ax.set_ylabel("Date")
ax.set_xlabel("Fare($)")
ax.invert_yaxis()
```

```python
## SCATTER PLOTS ## MATLAB

# Create a scatter plot
# You can use this code too: plt.plot(x_axis, y_axis, 'o

# Create a scatter plot
plt.scatter(y_axis, x_axis,color="red", label="Chicago")
plt.title("Pyber Fare by Month")
plt.ylabel("Fare($)")
plt.xlabel("Date")
plt.gca().invert_yaxis()
```

```python
## BUBBLE CHARTS ## MATLAB

plt.scatter(x_axis, y_axis, s=y_axis)

# Adjust size a little larger for easy readability
y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)

plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# Instead of using for loops and while loops, python can spot a pattern in this example below
# new_list = [expression for item in list if conditional]
# add the technique above to the plt.scatter() function
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
## OBJECT-ORIENTED SCATTER & BUUBLE ##

fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s = [i * 5 for i in y_axis], color="skyblue", label="Boston", edgecolors="black", linewidths=2)
ax.legend()
ax.set_ylabel("Fare($)")
ax.set_xlabel("Date")
```

```python
## PIE CHART ## MATLAB METHOD

plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
# The "explode" parameter will offset the indicated wedge by a fraction of the radius, 
# where "0" is zero distance from the center of the pie, and "1" is completely outside the diameter of the pie.
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct= '%.1f%%')
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", 
          "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

```python
## PIE CHART ## OOP

fig, ax = plt.subplots()
ax.pie(y_axis,labels=x_axis)
plt.show()
```
